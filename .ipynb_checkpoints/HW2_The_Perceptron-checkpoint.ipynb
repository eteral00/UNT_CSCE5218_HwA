### Build the Perceptron Model

You will need to complete some of the function definitions below.  DO NOT import any other libraries to complete this. 

````python
import math
import itertools
import re



# Corpus reader, all columns but the last one are coordinates;
#   the last column is the label
def read_data(file_name):
    f = open(file_name, 'r')

    data = []
    # Discard header line
    f.readline()
    for instance in f.readlines():
        if not re.search('\t', instance): continue
        instance = list(map(int, instance.strip().split('\t')))
        # Add a dummy input so that w0 becomes the bias
        instance = [-1] + instance
        data += [instance]
    return data


def dot_product(array1, array2):
    #Return dot product of array 1 and array 2
    # actually, list 1 & 2, not arrays 
    dotProduct = 0
    for num1, num2 in zip(array1, array2):
        dotProduct += (num1 * num2)
    return dotProduct 


def sigmoid(x):
    #Return outpout of sigmoid function on x
    eulerNum = math.e    
    return 1/(1 + (eulerNum**(-x) ) )


# The output of the model, which for the perceptron is 
# the sigmoid function applied to the dot product of 
# the instance and the weights
def output(weights, instance):
    #return the output of the model 
    return sigmoid( dot_product( list(weights), list(instance) ) )


# Predict the label of an instance; this is the definition of the perceptron
# you should output 1 if the output is >= 0.5 else output 0
def predict(weights, instance):
    #return the prediction of the model
    if output(weights, instance) >= 0.5:
        return 1
    else:
        return 0


# Accuracy = percent of correct predictions
def get_accuracy(weights, instances):
    # You do not to write code like this, but get used to it
    correct = sum([1 if predict(weights, instance) == instance[-1] else 0
                   for instance in instances])
    return correct * 100 / len(instances)


# Train a perceptron with instances and hyperparameters:
#       lr (learning rate) 
#       epochs
# The implementation comes from the definition of the perceptron
#
# Training consists on fitting the parameters which are the weights
# that's the only thing training is responsible to fit
# (recall that w0 is the bias, and w1..wn are the weights for each coordinate)
#
# Hyperparameters (lr and epochs) are given to the training algorithm
# We are updating weights in the opposite direction of the gradient of the error,
# so with a "decent" lr we are guaranteed to reduce the error after each iteration.
def train_perceptron(instances, lr, epochs):

    # initialize the weights vector to all zeros (with length of weights vector = number of features)
    weights = [0] * (len(instances[0])-1)

    # loop n times, with n = input epochs
    for _ in range(epochs):
        # going through all input instances and train with each instance
        for instance in instances:
            # compute the output, i.e. the predicted label/number, using the features vector & current weights vector
            in_value = dot_product(weights, instance[0:-1]) # omitting the last element in vector instance, which is the label
            output = sigmoid(in_value)
            error = instance[-1] - output
            # adjust each of the weights in the weights vector 
            # using the previously computed output & error, and input learning rate
            for i in range(0, len(weights)):
                weights[i] += lr * error * output * (1-output) * instance[i]    
    # finally, return the final weights vector
    return weights
````

## Run it

````python
instances_tr = read_data("train.dat")
instances_te = read_data("test.dat")
lr = 0.005
epochs = 5
weights = train_perceptron(instances_tr, lr, epochs)
accuracy = get_accuracy(weights, instances_te)
print(f"#tr: {len(instances_tr):3}, epochs: {epochs:3}, learning rate: {lr:.3f}; "
      f"Accuracy (test, {len(instances_te)} instances): {accuracy:.1f}")
````

### Question 1

In `train_perceptron(instances, lr, epochs)`, we have the following code:
```
in_value = dot_product(weights, instance)
output = sigmoid(in_value)
error = instance[-1] - output
```

Why don't we have the following code snippet instead?
```
output = predict(weights, instance)
error = instance[-1] - output
```

#### Answers (text only):

The first code set gives more accurate & "differentiable" values for output & error, thus leading to better adjustment of the weights (which needs to use values output & error) 
Also, the first set reflect the meaning of using the sigmoid function (or any activation function other than the binary step function, for that matter). Whereas the second code set uses predict method to "round" the output of the sigmoid function into the class-labels (1|0), essentially making the use of sigmoid meaningless, and turning things back to the ways of the binary step function - where both output and error can only be 1|0 and not differentiable -> making weights also only adjustable between a limited number of values.

````python
instances_tr = read_data("train.dat")
instances_te = read_data("test.dat")
tr_percent = [5, 10, 25, 50, 75, 100] # percent of the training dataset to train with
num_epochs = [5, 10, 20, 50, 100]     # number of epochs
lr_array = [0.005, 0.01, 0.05]        # learning rate

allAccuracies = []
for lr in lr_array:
    for tr_size in tr_percent:
        for epochs in num_epochs:
            size =  round(len(instances_tr)*tr_size/100)
            pre_instances = instances_tr[0:size]
            weights = train_perceptron(pre_instances, lr, epochs)
            accuracy = get_accuracy(weights, instances_te)
            allAccuracies += [accuracy]
            print(f"#tr: {len(pre_instances):0}, epochs: {epochs:3}, learning rate: {lr:.3f}; "
                  f"Accuracy (test, {len(instances_te)} instances): {accuracy:.1f}")
````

#### Answers (text here, code below (if applicable) ):
    ** In general, the trend(s) here is(are)
        More training data -> higher accurary
        More training epoch -> higher accuracy (to a certain number of epoch)
        Higher learning rate -> higher accuracy or actually faster increment of accuracy when adding more training data/epochs (not absolute/always increase, sometimes, higher learning rate does decrease the accuracy)
    ** The plots below show those trends in visual form.
        
- A/ In our particular case, we do not need to train with all the training dataset to get the highest accuracy with the test dataset. For instance, there is this case, [#tr: 300, epochs: 100, learning rate: 0.010; Accuracy (test, 100 instances): 80.0]. Here, we only used 75% of the training set to get 80% accuracy with the test set, which is higher than multiple cases with 100% of training set. In general, however, the trend is <more training data -> higher test accuracy, given all other things the same>. Again, the plots below show this trend among others.

- B/ In this case, the first run used less training data but used a higher learning rate (0.05, compared to 0.005 of the 2nd run, so it was 10x higher). So, the perceptron learned faster and moved faster toward the bottom of the "error descent". The real issue with that higher learning rate hasn't showed up yet in that case, and usually would only shows up when the perceptron is fed a lot more training data. When it is given more data and gets closer to the error bottom, the higher learning rate may make it "jump" around the bottom instead of descending (which is called "unstability")
    There is also another possibility (though rare) where EVEN WITH ALL OTHER HYPERPARAMETERS STAYING THE SAME, a run with less training data would get higher test accuracy than a run with more data. That is, if the smaller part of train dataset that is used in the first run better resembles the data in the test set and the additional part in the second run actually only "dilutes" the model. Such cases would happen more when we have small train & test set and we select data for train/test in sequence (instead of randomly select & shuffle). And These cases would hardly occur as the size of train & test sets goes up or if we use more random-procedure in selecting data for train & test.

- C/ Theoretically, we can get higher accuracy with more hyperparameters. Yes, sure we can. The other side of the equation is, it will also make our model more complicated, more specialized, less generalizable, and increases the chance of overfitting. This is similar to the case of <producing clothes for a population vs. tailoring for a single person>. We can add a lot of hyperparmeters & select values for each of them to tailor our model to fit better to this particular given test set, which at the same time may make our model perform worse with other datasets.

- D/ Short answer: No, it is not always worth training for more epochs. 
    Longer answer: There is a diminishing return for training epochs, as well as a plateau level (converge point). However, the extent of those effects also depends on other hyperparameters such as the size of input dataset & learning rate. In general, the first few epochs usually increase accuracy greatly/rapidly. That rate of increase also goes down quickly and the later epochs can only increase accuracy little by little. After a certain number of epochs, the accuracy would plateau and may also revolve (go up, down, up, down, etc.) around that plateau level. So, adding more training epochs after that point is useless and wasteful.

````python
import matplotlib.pyplot as myPlot


def plot_accuracy_by_dimension (dimension, dimensionLabel, accuraciesByDimension):
    myPlot.plot(dimension, accuraciesByDimension)
    myPlot.ylim([min(accuraciesByDimension)-1, max(accuraciesByDimension)+1])
    myPlot.xlabel(dimensionLabel)
    myPlot.ylabel("Accuracy")
    myPlot.title("Average Accuracy by " + dimensionLabel)
    myPlot.show()
````

````python
dimensionLength_lr = len(lr_array)
numOfElement_lr = int(len(allAccuracies)/dimensionLength_lr)
accuraciesByLr = []
accuracy = 0
for idx in range(dimensionLength_lr):    
    start = int(idx * numOfElement_lr)
    end = int((idx+1) * numOfElement_lr)
    accuracy = sum(allAccuracies[start:end]) / numOfElement_lr
    accuraciesByLr += [accuracy]

dimensionLabel = "Learning Rate"
plot_accuracy_by_dimension(lr_array, dimensionLabel, accuraciesByLr)


dimensionLength_tr = len(tr_percent)
numOfElement_tr = int(len(allAccuracies)/dimensionLength_tr)
accuraciesByTrPercent = []

dimensionLength_ne = len(num_epochs)
numOfElement_ne = int(len(allAccuracies)/dimensionLength_ne)
accuraciesByNumEpochs = []
for idx in range(dimensionLength_tr):
    accuracy = 0
    for idx2 in range(dimensionLength_lr):
        start = int( (idx2 * numOfElement_lr) + (idx * dimensionLength_ne) )
        end = start + dimensionLength_ne
        accuracy += sum(allAccuracies[start:end])
    accuracy = accuracy / numOfElement_tr
    accuraciesByTrPercent += [accuracy]

dimensionLabel = "Percentage of Training Set"
plot_accuracy_by_dimension(tr_percent, dimensionLabel, accuraciesByTrPercent)


for idx in range(dimensionLength_ne):
    accuracy = 0
    for idx2 in range(numOfElement_ne):
        accuracy += allAccuracies[idx + (idx2 * dimensionLength_ne)]
    accuracy = accuracy / numOfElement_ne
    accuraciesByNumEpochs += [accuracy]
    
dimensionLabel = "Number of Epochs"
plot_accuracy_by_dimension(num_epochs, dimensionLabel, accuraciesByNumEpochs)
````
